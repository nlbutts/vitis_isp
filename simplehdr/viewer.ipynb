```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 12]
```

```python
def print_stats(num, img):
    maxv = np.max(img)
    minv = np.min(img)
    print(f'Num: {num} Type: {img.dtype} Min: {minv} Max: {maxv}')

def view(img):
    img = img.astype('float32')
    img /= np.max(img)
    plt.imshow(img, cmap='gray')
```

```python
img = cv.imread('CaptureLinear.png', -1)
Hp = cv.imread('Hp.png', -1)
temp1 = cv.imread('temp1.png', -1)
Ibip = cv.imread('ibip.png', -1)
Ap = cv.imread('Ap.png', -1)
Iga = cv.imread('Iga_float.png', -1)
```
