```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 12]
```

```python
def show(img):
    flip = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    flip = flip.astype('float32')
    flip /= np.max(flip)
    plt.imshow(flip)
```

```python
x = np.fromfile('ar0231_b.raw', 'uint16')
x *= 16
print(len(x) / 1928)

y = np.reshape(x, (1208, 1928))
plt.imshow(y)
cv.imwrite('ar0231_b.png', y)
```
